```python
df.replace("S",1,inplace =True)#To replace the values and inplace to make it permanent
df.replace("C",2,inplace =True)
df.replace("Q",3,inplace =True)
df
```

```python
df["Embarked"].head()
```

```python
df = df.drop("Name", axis=1)
df
df.head()
```

```python
df["Age"]=df["Age"].astype("int")
df["Fare"]=df["Fare"].astype("int")
df.replace("male",0,inplace =True)#To replace the values and inplace to make it permanent
df.replace("female",5,inplace =True)
df
```

```python
df["Ticket"].value_counts()
```

```python
def replace_non_numeric_with_1000(value):
    try:
        # Try to convert the value to a numeric (int or float)
        return int(value)
    except :
        # If the conversion fails, return 1000
        return 1000
df["Ticket"] = df["Ticket"].apply(replace_non_numeric_with_1000)

print(type(df["Ticket"]))
df.info()
df["Ticket"].value_counts()
```

```python
x = df.drop("Embarked", axis=1)#specifically col
```

```python
x
```

```python
y = df["Embarked"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=0.65,random_state=20)#x_train,test and y_train,test please change at karna
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,Y_train)#LogisticRegression() function call kar raha hai
```

```python
y_pred=log.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,f1_score,precision_score,accuracy_score,recall_score
```

```python
conf=confusion_matrix(Y_test,y_pred)
print(conf)
```

```python
f1=f1_score(Y_test,y_pred,average='weighted')
print(f1)
```

```python
re=recall_score(Y_test,y_pred,average='weighted')
print(re)
```

```python
acc=accuracy_score(Y_test,y_pred)
print(acc)
```
